# Scatter Plot

```python

a=[1,2,3,4,5,6,7]
b=[2,3,1,4,5,3,6]
plt.xlabel("Dates",fontsize=20)
plt.ylabel("Figures",fontsize=20)
plt.title("Scatter plot",fontsize=20)
plt.scatter(a,b ,color="orange",s=500,edgecolor="red",linewidth=5) #linewidth edgecolor ko zaida kr deta hai
plt.show()
```

# Multiple Colors in Scatter Plot

```python
a=[1,2,3,4,5,6,7]
b=[2,3,1,4,5,3,6]
plt.xlabel("Dates",fontsize=20)
plt.ylabel("Figures",fontsize=20)
plt.title("Scatter plot",fontsize=20)
color=["orange","green","yellow","purple","red","grey","pink"]
plt.scatter(a,b ,c=color,s=500,edgecolor="GREEN") #Edge color circle k side ko color krta hai
plt.show()
```

# Multiple Sizes in Scatter Plot

```python
a=[1,2,3,4,5,6,7]
b=[2,3,1,4,5,3,6]
plt.xlabel("Dates",fontsize=20)
plt.ylabel("Figures",fontsize=20)
plt.title("Scatter plot",fontsize=20)
color=["orange","green","yellow","purple","red","grey","pink"]
sizes=[120,210,190,200,150,130,160]
plt.scatter(a,b ,c=color,marker="*",s=sizes, alpha=0.9)# alpha stars k color ko blur krny k lai use hota hai
plt.show()
```

```python
import matplotlib.pyplot as plt

a = [1, 2, 3, 4, 5, 6, 7]
b = [2, 3, 1, 4, 5, 3, 6]
colors = [20, 30, 40, 50, 60, 70, 80]  # this will control color intensity

plt.xlabel("Dates", fontsize=20)
plt.ylabel("Figures", fontsize=20)
plt.title("Scatter plot", fontsize=20)

plt.scatter(a, b, c=colors, s=500, cmap="viridis")
plt.colorbar()  # shows color legend
plt.show()
```

# Histogram Plot

```python
import matplotlib.pyplot as plt
marks=[12,13,56,15,16,45,18,60,20,43,22,23,24,89,26,27,90,100,67]
l=[10,15,25,35,40]
plt.hist(marks,color="green",edgecolor="red",bins=5,cumulative=-1,bottom=30,histtype="step",orientation="horizontal",log="True",label="python")
plt.legend() # label show hota h jesay python show ho rha hai
plt.title("histogram Plot")
plt.show()
```

# Pie Plot

```python
import matplotlib.pyplot as plt
a=[10,20,30,40,50]
b=['C','C++','Java','Javascript','Python']
ex=[0.04,0.0,0.05,0.0,0.1]
c=['orange','red','green','yellow','grey']
plt.pie(a,labels=b,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.9, labeldistance=1.1) #explode selected attribute ko apni jagah sy move kr deta hai
plt.show()
```

# Stem Plot

```python
import matplotlib.pyplot as plt
a = [12, 15, 17, 19, 22]
b = [50, 75, 100, 135, 170]

plt.stem(a, b, linefmt=":", markerfmt="ro", basefmt="yellow", label="Python",orientation="horizontal")
plt.legend()
plt.show()
```

# Box Vs Whisker Plot

```python
import matplotlib.pyplot as plt

X = [10, 20, 30, 40, 50, 60]

plt.boxplot([X], widths=0.1, labels=["Python"], patch_artist=True,showmeans=True,sym="g+",boxprops=dict(color="red"),capprops=dict(color="red"),whiskerprops=dict(color="red"))
plt.title("Box Plot")
plt.xlabel("Digit")
plt.ylabel("Values")
plt.show()
```

# Multiple Box Plots

```python
import matplotlib.pyplot as plt

X = [10, 20, 30, 40, 50, 60]
x1 = [10, 20, 40, 50, 30, 60, 90]
x2=  [12, 35, 45, 55, 75, 85, 95]
y=[X,x1,x2]

plt.boxplot(y, widths=0.3, labels=["Python","C++","C#"], patch_artist=True, showmeans=True, sym="g+",
            boxprops=dict(color="red"), capprops=dict(color="red"), whiskerprops=dict(color="red"))

plt.title("Box Plot")
plt.xlabel("Digit")
plt.ylabel("Values")
plt.show()
```
